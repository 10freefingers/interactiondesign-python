## Basic time series

  * usually: x axis = time
  * y axis = values/samples/measurements
  * even spacing of x axis

```python
# list of 100 evenly spaced points from 0 to 10
x = linspace(0,10,100)
# apply function to each point
plot(x, sin(x*2))
```

## Noisy data

Two types of noise:
  * process noise - physically present in the measured process, e.g. hand tremor
  * measurement noise - occurs in measuring equipment, e.g. thermal noise

```python
# list of 100 random points between -0.25 and +0.25
noise = (random.random(100)-0.5)/2
plot(x, sin(x*2))
plot(x, sin(x*2)+noise)
```

## Reminder: mean/standard deviation

  * N values, $x_i, i = 1...N$
  * mean: $\mu = \frac{1}{N} \sum_{i=1}^{N} x_i$
  * standard deviation: $\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N}(x_i - \mu)^2}$
  
  * important: two different commonly used estimators for standard deviation
    * difference

```python
print std(sin(x)), mean(sin(x))
```

```python
print std(sin(x)+noise), mean(sin(x)+noise)
```

## Filter kernels

  * filter "kernel": second function which is multiplied "on top of" original time series
  * simple example: square filter = moving average

```python
kernel = [ 0 for i in range(0,100) ]
for i in range(45,55):
    kernel[i] = 0.1
plot(x, kernel)
ylim(-0.05,0.15)
```

## Convolution

  * convolution: point-wise multiplication of two functions
  * definition: $y[i] = \sum_{j=0}^{M-1} h[j] * x[i-j]$
  
  ![convolution](http://www.dspguide.com/graphics/F_6_8.gif)

```python
plot(x, kernel)
plot(x, sin(x*2))
plot(x, convolve(sin(x*2)+noise,kernel,mode='same'))
ylim(-1.2,1.2)
```

```python
for i in range(25,75):
    kernel[i] = 1.0/50
    
plot(x, kernel)
plot(x, sin(x*2))
plot(x, convolve(sin(x*2)+noise,kernel,mode='same'))
```

## Fourier Transformation

  * DFT - discrete Fourier transformation: transform digital data from amplitude domain into frequency domain
  * FFT - fast Fourier transformation: typical implementation of DFT
  * see also: http://glowingpython.blogspot.de/2011/08/how-to-plot-frequency-spectrum-with.html

```python
def plotSpectrum(y,Fs):
 """
 Plots a Single-Sided Amplitude Spectrum of y(t)
 """
 n = len(y) # length of the signal
 k = arange(n)
 T = n/Fs
 frq = k/T # two sides frequency range
 frq = frq[range(n/2)] # one side frequency range

 Y = fft.fft(y)/n # fft computing and normalization
 Y = Y[range(n/2)]
 
 plot(frq,abs(Y),'r') # plotting the spectrum
 xlabel('Freq (Hz)')
 ylabel('|Y(freq)|')

Fs = 150.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = arange(0,1,Ts) # time vector

ff = 5.5;   # frequency of the signal
y = sin(2*pi*ff*t)

subplot(2,1,1)
plot(t,y)
xlabel('Time')
ylabel('Amplitude')
subplot(2,1,2)
plotSpectrum(y,Fs)
show()
```
